```python
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
Low = []
moderate = []
High = []
for study, entertainment, sleep in time_data:
    if study < 3:
        Low.append(study)
    elif 3 <= study <= 5:
        moderate.append(study)
    else:
        High.append(study)

print("Low:", Low)
print("Moderate:", moderate)
print("High:", High)
```

```python
study_hours = [s for s, _, _ in time_data]
entertainment_hours = [e for _, e, _ in time_data]
sleep_hours = [sl for _, _, sl in time_data]

def average(nums):
    return sum(nums) / len(nums)

print("Average study:", round(average(study_hours), 2))
print("Average entertainment:", round(average(entertainment_hours), 2))
print("Average sleep:", round(average(sleep_hours), 2))
```

```python
def factorial(n):
    """
    Returns n! for non-negative integer n.
    Raises ValueError for negative inputs.
    """
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(6))
```

```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

nested = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Sum of nested list:", sum_nested_list(nested))
```

```python
def get_dir_size(directory):
    size = 0
    for item, value in directory.items():
        if isinstance(value, dict):
            size += get_dir_size(value)
        else:
            size += value
    return size

file_system = {
    "file1": 200,
    "folder1": {
        "file2": 300,
        "file3": 400
    },
    "folder2": {
        "file4": 100
    }
}
print("Total directory size:", get_dir_size(file_system))
```

```python
def coin_change(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

print("Min coins for 11:", coin_change([1, 2, 5], 11))
```

```python
def lcs(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

print("LCS length:", lcs("abcde", "ace"))
```

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([5, 6])

print("Inverse:\n", np.linalg.inv(A))
print("Determinant:", np.linalg.det(A))
print("Frobenius Norm:", np.linalg.norm(A, 'fro'))
print("2-Norm (Euclidean):", np.linalg.norm(A))
print("Solution x:", np.linalg.solve(A, B))
```
